```python
base1 = Path("/mnt/dwango/hiroshiba/become-yukarin/dataset/hiho-wave/hiho-pause-atr503-subset/").expanduser()
base2 = Path("/mnt/dwango/hiroshiba/become-yukarin/dataset/yukari-wave/yukari-pause-atr503-subset/").expanduser()

base_model = Path('/mnt/dwango/hiroshiba/become-yukarin/cbhg-pause-notopdb-noise001/')
config = create_config(base_model / 'config.json')
voice_changer = VoiceChanger(config, base_model / 'predictor_100000.npz')
```

```python
path1 = str(list(base1.glob('*A01*'))[0])
wave, sr = librosa.load(path1, sr=config.dataset.param.voice_param.sample_rate)
Audio(wave, rate=sr)
```

```python
path2 = str(list(base2.glob('*A01*'))[0])
wave, sr = librosa.load(path2, sr=config.dataset.param.voice_param.sample_rate)
Audio(wave, rate=sr)
```

```python
def normarize_wav(wave: Wave):
    y = wave.wave
    y *= 32767 / max(0.01, numpy.max(numpy.abs(y)))
    y[y > 32767] = 32767
    y[y < -32767] = -32767
    return y


wave = voice_changer(path1)
y = normarize_wav(wave)
Audio(y, rate=config.dataset.param.voice_param.sample_rate)
```
